```python
import string
import unicodedata
```

```python
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
```

```python
for file in txtarray:
     fileObj= open('/content/drive/MyDrive/PulaMuvula/'+file, 'r')
     text = fileObj.read()
     results = text.split('.')


     with open('/content/drive/MyDrive/PulaMvulaFinalProcessed/'+file, "w") as file_:
          special_characters=['@','#','$','*','&', ':', '=', '(', ')', '/', '\\', '\uf03c', '1','0','2','3','4','5','6','7','8','9']
          for result in results:
              for i in special_characters:
                  result=result.replace(i,"")
              file_.write(strip_accents(result).translate(str.maketrans('', '', string.punctuation)))
```
